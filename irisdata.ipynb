```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dir(load_iris())
df = pd.DataFrame(load_iris().data,columns=load_iris().feature_names)
df.head()
```

```python
df['label'] = load_iris().target
df.head(100)
```

```python
f,a = plt.subplots(nrows=1,ncols=4)
# plt.figure(figsize=(5,15))
f.tight_layout()
a[0].scatter(df['sepal length (cm)'],df['label'])
a[1].scatter(df['sepal width (cm)'],df['label'])
a[2].scatter(df['petal length (cm)'],df['label'])
a[3].scatter(df['petal width (cm)'],df['label'])
```

```python
features = df.iloc[:,:-1].values
label = df.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,label,test_size=0.25,random_state=21)
```

```python
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
svc = SVC()
mnbyes = GaussianNB()
```

```python
decision_tree.fit(x_train,y_train)
random_forest.fit(x_train,y_train)
svc.fit(x_train,y_train)
mnbyes.fit(x_train,y_train)
```

```python
pred1 = decision_tree.predict(x_test)
pred2 = random_forest.predict(x_test)
pred3 = svc.predict(x_test)
pred4 = mnbyes.predict(x_test) 
```

```python
print('the accuray score of decision tree is {}'.format(accuracy_score(y_test,pred1)))
print('the accuray score of random_forest is {}'.format(accuracy_score(y_test,pred2)))
print('the accuray score of svc is {}'.format(accuracy_score(y_test,pred3)))
print('the accuray score of GaussianNB  is {}'.format(accuracy_score(y_test,pred4)))
```
